```python
 import os
import re

def load_words_with_weight(file_path):
    match = re.search(r"（(.*?)）", file_path)
    if match:
        weight = float(match.group(1))
    else:
        weight = 1.0

    words = []
    # 尝试使用不同的编码格式读取文件
    for encoding in ('utf-8', 'gbk', 'latin1'):
        try:
            with open(file_path, 'r', encoding=encoding) as f:
                words = [line.strip() for line in f.readlines() if line.strip()]
            break  # 如果成功读取，则跳出循环
        except UnicodeDecodeError:
            continue  # 如果出现编码错误，则尝试下一个编码

    if not words:  # 如果所有尝试都失败了
        raise UnicodeDecodeError(f"Cannot decode file {file_path} with tried encodings.")

    return words, weight


def build_word_dict_from_files(*files):
    word_dict = {}
    for file_path in files:
        words, weight = load_words_with_weight(file_path)
        for word in words:
            word_dict[word] = weight
    return word_dict

# 假设文件路径列表
degree_files = [
    "/Users/kkyang/Desktop/否定（-0.8）.txt",
    "/Users/kkyang/Desktop/超（0.9）.txt",
    "/Users/kkyang/Desktop/很（0.7）.txt",
    "/Users/kkyang/Desktop/较（0.5）.txt",
    "/Users/kkyang/Desktop/欠（-0.5）.txt",
    "/Users/kkyang/Desktop/稍（0.3）.txt",
    "/Users/kkyang/Desktop/最（0.9）.txt",
]

# 构建程度副词字典
degree_dict = build_word_dict_from_files(*degree_files)

# 打印结果以验证
print(degree_dict)
```

```python
def emotion_caculate(text, emotion_dict, degree_dict, negation_dict):
    wordlist = jieba.lcut(text)
    score = 0

    # 初始化加权值和否定状态
    last_degree_weight = 1
    last_negation_weight = 1
    
    for word in wordlist:
        # 检查并应用程度副词的加权值
        if word in degree_dict:
            last_degree_weight = degree_dict[word]
            continue
        # 检查并应用否定词的加权值
        if word in negation_dict:
            last_negation_weight *= negation_dict[word]  # 可能连续出现否定词
            continue

        # 应用情感词
        if word in emotion_dict:
            intensity, polarity = emotion_dict[word]
            adjusted_intensity = intensity * last_degree_weight * last_negation_weight
            score += adjusted_intensity * polarity
            
            # 重置加权值和否定状态
            last_degree_weight = 1
            last_negation_weight = 1
    
    return score
```
